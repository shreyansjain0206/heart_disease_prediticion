```python
from sklearn.preprocessing import LabelEncoder
label= LabelEncoder()
category_col=df.select_dtypes(include=['object']).columns
```

```python

for col in category_col:
  if len(df[col].unique())<=2:
    df[col]=label.fit_transform(df[col])
  else:
    df=pd.get_dummies(df,columns=[col])
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

```

```python
Y=df.iloc[:,0]
Y.head()
```

```python
X=df.iloc[:,1:]
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaled_x = scaler.transform(X_train)
scaled_X = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
rf_classifier.fit(scaled_x, Y_train)
```

```python
y_predictions = rf_classifier.predict(scaled_X)
```

```python
tr_accuracy = rf_classifier.score(scaled_x, Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(Y_test, y_predictions)
```

```python
print("Train Accuracy:", f'{round(tr_accuracy, 4) * 100}%')
print("Test Accuracy:", f'{round(accuracy, 4) * 100}%')
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
cm = confusion_matrix(Y_test, y_predictions)
print('confusion Matrix')
print(cm)
```

```python
cm = confusion_matrix(Y_test, y_predictions)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm)
cm_display.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
cm = confusion_matrix(Y_test, y_predictions)

plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
```
